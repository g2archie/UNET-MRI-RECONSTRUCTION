```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('UNet2D2D_SSIM_loss_no_regularization_plot_data.pkl', 'rb') as f:
    unet_2d2d = pickle.load(f)
```

```python
with open('UNet2D1D_SSIM_loss_no_regularization_plot_data.pkl', 'rb') as f:
    unet_2d1d = pickle.load(f)
```

```python
with open('UNet3D_SSIM_loss_no_regularization_plot_data.pkl', 'rb') as f:
    unet_3d = pickle.load(f)
```

```python
with open('UNet3D_old_SSIM_loss_no_regularization_plot_data.pkl', 'rb') as f:
    unet_3d_old = pickle.load(f)
```

```python
task_name = 'best_images'
plt.figure(1, figsize=(20, 20))

plt.subplot(4, 3, 1)
plt.axis('off')
plt.title('UNet2D2D - Input')
plt.imshow(unet_2d2d[task_name][0], cmap='gray')

plt.subplot(4, 3, 2)
plt.axis('off')
plt.title('UNet2D2D - Reconstructed Image')
plt.imshow(unet_2d2d[task_name][1], cmap='gray')

plt.subplot(4, 3, 3)
plt.axis('off')
plt.title('UNet2D2D - Ground Truth')
plt.imshow(unet_2d2d[task_name][2], cmap='gray')

plt.subplot(4, 3, 4)
plt.axis('off')
plt.title('UNet2D1D - Input')
plt.imshow(unet_2d1d[task_name][0], cmap='gray')

plt.subplot(4, 3, 5)
plt.axis('off')
plt.title('UNet2D1D - Reconstructed Image')
plt.imshow(unet_2d1d[task_name][1], cmap='gray')

plt.subplot(4, 3, 6)
plt.axis('off')
plt.title('UNet2D1D - Ground Truth')
plt.imshow(unet_2d1d[task_name][2], cmap='gray')

plt.subplot(4, 3, 7)
plt.axis('off')
plt.title('UNet3D - Input')
plt.imshow(unet_3d[task_name][0], cmap='gray')

plt.subplot(4, 3, 8)
plt.axis('off')
plt.title('UNet3D - Reconstructed Image')
plt.imshow(unet_3d[task_name][1], cmap='gray')

plt.subplot(4, 3, 9)
plt.axis('off')
plt.title('UNet3D - Ground Truth')
plt.imshow(unet_3d[task_name][2], cmap='gray')
plt.subplot(4, 3, 10)
plt.axis('off')
plt.title('UNet3D_old - Input')
plt.imshow(unet_3d_old[task_name][0], cmap='gray')

plt.subplot(4, 3, 11)
plt.axis('off')
plt.title('UNet3D_old - Reconstructed Image')
plt.imshow(unet_3d_old[task_name][1], cmap='gray')

plt.subplot(4, 3, 12)
plt.axis('off')
plt.title('UNet3D_old - Ground Truth')
plt.imshow(unet_3d_old[task_name][2], cmap='gray')
plt.savefig('Best_reconstructed_images.jpg', bbox_inches='tight')
```
