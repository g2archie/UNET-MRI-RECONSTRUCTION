```python
import pickle
import math
from collections import OrderedDict 
```

```python
with open('laptop_result.pkl', 'rb') as f:
    laptop_result = pickle.load(f)
with open('desktop_result.pkl', 'rb') as f:
    desktop_result = pickle.load(f)
with open('hannover_result.pkl', 'rb') as f:
    hannover_result = pickle.load(f)
```

```python
combined_result = [*desktop_result, *laptop_result, *hannover_result]
```

```python
len(combined_result)
```

```python
combined_result.sort(key=lambda r:r[1][3], reverse=True)
```

```python
combined_result[:2]
```

```python
combined_result_dict = OrderedDict()
```

```python
for k, v in combined_result:
    if k not in combined_result_dict and not math.isnan(v[1]) :
        combined_result_dict[k] = v
```

```python
combined_result_dict
```

```python
with open('combined_evaluation_result.pkl', 'wb') as f:
    pickle.dump(combined_result_dict, f)
```
